# 태그 확인

```python
tags = df['key'].apply(lambda x:x!=[])
```

```python
df.info()
```

```python
# 문자열을 리스트로 변환하는 함수 정의
def string_to_list(s):
    try:
        return ast.literal_eval(s)
    except ValueError:
        return []  # 또는 적절한 기본값 반환
```

```python
df['key'] = df['key'].apply(string_to_list)
```

```python
exploded_df = df.explode('key')
exploded_df
```

```python
check = exploded_df.key.value_counts()

# seaborn을 사용한 시각화
sns.barplot(x=check.index, y=check.values)

# matplotlib 기본 설정으로 그래프 꾸미기
plt.xlabel('Key')
plt.ylabel('Counts')
plt.title('Frequency of Each Key')
plt.xticks(rotation=45)  # x축 레이블 회전
plt.show()
```
